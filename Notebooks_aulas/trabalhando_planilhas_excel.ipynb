# Trabalhando com Planilhas do Excel

```python
# Importando biblioteca pandas
import pandas as pd
```

```python
# Leitura dos arquivos
df_1 = pd.read_excel("Aracaju.xlsx")
df_2 = pd.read_excel("Fortaleza.xlsx")
df_3 = pd.read_excel("Natal.xlsx")
df_4 = pd.read_excel("Recife.xlsx")
df_5 = pd.read_excel("Salvador.xlsx")
```

```python
# Juntando todos os arquivos
df = pd.concat([df_1,df_2,df_3,df_4,df_5])
```

```python
# Exibindo as cinco primeiras linhas
df.head()
```

```python
# Retornar valores unicos da coluna cidades
df["Cidade"].unique()
```

```python
# Exibindo as cinco ultimas linhas
df.tail()
```

```python
# Para selecionar amostra do dataset
df.sample(5)
```

```python
# Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
# Alterando o tipo de dado da coluna LojaID de int para object
df["LojaID"] = df["LojaID"].astype("object")
df.dtypes
```

Trabalhando com valores tipo NULL (nulos)


```python
# Consultando e retornando o somatório dos valores NULL
df.isnull().sum()
```

```python
# Substituindo linhas com valores NULL em memória com média ex: 
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
# Substituindo linhas com valores NULL por zero em memoria
df["Vendas"].fillna(0, inplace=True)
```

```python
# Removendo linhas com valores NULL em memoria
df.dropna(inplace=True)
```

```python
# Removendo linhas com valores NULL em memoria com base em apenas 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
# Removendo linhas com valores NULL em memoria com base em todas as colunas
df.dropna(how="all", inplace=True)
```

Criando novas colunas

```python
# Criando coluna receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
# Retorando a maior receita
df["Receita"].max()
```

```python
# Retorando a menor receita
df["Receita"].min()
```

```python
# Retorando as 3 lojas com maiores receitas
df.nlargest(3, "Receita")
```

```python
# Retorando as 3 lojas com menores receitas
df.nsmallest(3, "Receita")
```

```python
# Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
# Ordenando dataset
df.sort_values("Receita", ascending=False).head(10)
```
